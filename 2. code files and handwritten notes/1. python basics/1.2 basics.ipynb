### 3. Write a Python program to check if a number is positive, negative, or zero.

### 3. Write a Python program to find the largest of three numbers.

```python
# taking 3 numbers as input and splitting them with split() function with ',' as the parameter
# stripping the numbers of any spaces with strip() and converting them into float 
# list comprehension to form a list of float numbers
# convert them into a tuple and unpack manually

n1,n2,n3 = tuple([float(x.strip()) for x in input("Please enter 3 numbers seperated by commas").split(',')])
print(max(n1,n2,n3))
```

### 5. Write a Python program to calculate the factorial of a number.

```python

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

num = int(input("Enter a number: "))
print(f"The factorial of {num} is {factorial(num)}")
```

## Basic Operators (Arithmetic, Comparison, Logical)

### 11. Write a Python program to perform arithmetic operations: addition, subtraction, multiplication, division, modulus, absolute division.

```python

a = 13
b = 6

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Modulus: {a} % {b} = {a % b}")      # to find the remainder
print(f"Division: {a} // {b} = {a // b}")   # absolute quotient after division, or floored division
```

### 16. Write a Python program to find the sum of the first n natural numbers.

```python
# the sum of first n natural numbers is n(n+1)/2

n = int(input("Enter a number: "))
sum_n = (n * (n + 1)) // 2
print(f"The sum of the first {n} natural numbers is {sum_n}")
```

### 17.   Write a Python program to check if a year is a leap year.

```python
year = int(input("Enter a year: "))
print(f"{year} is a leap year." if year % 4 == 0 else f"{year} is not a leap year.")
```

### 18. Write a Python program to reverse a string.

```python

# - slicing

string = input("Enter a string: ")
reversed_string = string[::-1]
print(f"The reversed string is: {reversed_string}")
```

### 19. Write a Python program to check if a string is a palindrome.

```python

string = input("Enter a string: ")
if string == string[::-1]:
    print(f"{string} is a palindrome.")
else:
    print(f"{string} is not a palindrome.")
```

### 20. Write a Python program to sort a list of numbers in ascending order.

```python
# 2 ways - list.sort(reverse=False) or lst = sorted(list)
numbers = [float(x) for x in input("Enter numbers separated by space: ").split(' ')]
numbers.sort()
print(f"Sorted list: {numbers}")
```
