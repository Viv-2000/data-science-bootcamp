### 6. List Methods

Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

```python
# random.randint() returns a random number from the given range
# random.randint(a,b) randomly chooses from the range a to b with a and b inclusive
# random.choice returns a random choice from the given choices
import random
random_numbers = [random.randint(1, 20) for _ in range(15)]
print(f"Original list: {random_numbers}")

random_numbers.sort()
print(f"Sorted in ascending order: {random_numbers}")

# option 1 is sorted_numbers = sorted(random_numbers)
random_numbers.sort(reverse=True)
print(f"Sorted in descending order: {random_numbers}")
# option 2 is sorted_numbers_desc = sorted(random_numbers, reverse=True)
# a set does not allow duplicates
unique_numbers = list(set(random_numbers))
print(f"List with duplicates removed: {unique_numbers}")
```

### 7. Nested Lists

Create a nested list representing a 4x6 matrix and print the matrix. Access and print the element at the second row and third column.

```python
import random
# random.randint(a,b) randomly chooses from the range a to b with a and b inclusive
matrix = [[random.randint(0,9) for _ in range(6)] for _ in range (4)]
print("Matrix:")
for row in matrix:
    print(row)
print(f"Element at second row and third column: {matrix[1][2]}")
```

### 8. List of Dictionaries

Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

```python
### this is a very important example as it uses dictionaries
### sorted(data_struc, key= , reverse=)
### labmda as a key(parameter of sorted() function), uses the value of a key(in the dict) within a dictionary for sorti
students = [
{'name': 'A', 'score': 88},
{'name': 'B', 'score': 72},
{'name': 'C', 'score': 95},
{'name': 'D', 'score': 65},
{'name': 'E', 'score': 78}
]
sorted_students = sorted(students, key=lambda x: x['score'], reverse=True)
print("Sorted students by score in descending order:")
for student in sorted_students:
    print(student)
```

### 9. Matrix Transposition

Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

```python
## zip() only combines 2 lists of equal lengths
## option 2 -- interview gold
## * character unpacks the list into 3 individual lists
## zip() function zips the 3 lists into 3 individual tuples
## list comprehension converts all individual tuples into a list, forming a matrix

def transpose_matrix(matrix):
    transposed = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    return transposed

## option 2 -- interview gold
def t2(matrix):
    return [list(row) for row in (zip(*matrix))]
matrix = [[random.randint(0,9) for _ in range(3)] for _ in range (3)]
transposed = transpose_matrix(matrix)

print("Transposed matrices:")
for row in transposed:
    print(row)
print('\n',t2(matrix),'\n')

print(list(zip(*matrix))) ## returns the transpose but in the incorrect format
```

### 10. Flattening a Nested List

Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

```python
def flatten_list(nested_list):
    flat_list = [item for sublist in nested_list for item in sublist]
    return flat_list

matrix = [[random.randint(0,9) for _ in range(3)] for _ in range (4)]

flattened = flatten_list(matrix)
print("Original nested list:")
print(matrix)
print("Flattened list:")
print(flattened)
```

### 11. List Manipulation

Create a list of the first 10 positive integers. Remove the elements at indices 2, 4, and 6, and insert the element '99' at index 5. Print the modified list. Pop the last element, print the modified list. pop any 2nd and 3rd elements, print the final list.

```python
lst = list(range(1, 11))
print(f"Original list: {lst}")
del lst[6]
del lst[4]
del lst[2]
lst.insert(5, 99)
print(f"Modified list: {lst}")
lst.pop()
print(f"Modified list: {lst}")
## after popping the 2nd element, the 3rd element becomes the 2nd again, so we pop second element twice
lst.pop(1)
lst.pop(1)
print(f"Final list: {lst}")
```

### 13. List Reversal

Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

```python
## easiest method is to use list slicing
def reverse_list(lst):
    return lst[::-1]

original_list = [1, 2, 3, 4, 5]
reversed_list = reverse_list(original_list)
print(f"Original list: {original_list}")
print(f"Reversed list: {reversed_list}")
```

### 14. List Rotation  (VERY IMPORTANT)

Write a function that rotates a list by n positions. Print the original and rotated lists.

```python
def rotate_list(lst, n):
    return lst[n:] + lst[:n]

original_list = [1, 2, 3, 4, 5]
rotated_list = rotate_list(original_list, 2)
print(f"Original list: {original_list}")
print(f"Rotated list: {rotated_list}")
```

### 15. List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

```python
def list_intersection(lst1, lst2):
    return [x for x in lst1 if x in lst2]

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
intersection = list_intersection(list1, list2)
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Intersection: {intersection}")
```
